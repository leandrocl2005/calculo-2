## Derivadas parciais

Considere a função $$f(x,y) = -x^2 -y^2.$$

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np
import pandas as pd
from matplotlib import cm
from IPython.display import clear_output
from matplotlib.ticker import LinearLocator, FormatStrFormatter



for angle in range(30,330,5):
    fig = plt.figure(figsize=(8,8))
    ax = fig.gca(projection='3d')

    # Make data.
    X = np.arange(-2, 2, 0.01) 
    Y = np.arange(-2, 2, 0.01) 
    X, Y = np.meshgrid(X, Y)
    Z = -X**2 -Y**2

    # Plot the surface.
    surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                        linewidth=0, antialiased=False, alpha=0.3)
    

    # Customize the z axis.
    ax.set_zlim(-10, 0.5)
    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

    ax.view_init(30,angle)

    # Add a color bar which maps values to colors.
    fig.colorbar(surf, shrink=0.5, aspect=5)

    # configura o ângulo
    ax.view_init(30,angle)

    # mostra
    plt.tight_layout()
    plt.gca()
    clear_output(wait=True)
    plt.show()
```

O paraboloide circular representa o gráfico da função $f$.

Calcular $f_x(0,0)$ é o mesmo que calcular $g'(0)$ com $g(x) = f(x,0)$

Vamos mudar o parâmetros por questões didáticas: g(t) = f(t,0)

Fazendo $z = g(t)$, temos que $(x, y, z) = (t, 0, -t^2)$. Vejamos onde essa curva está no gráfico:

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np
import pandas as pd
from matplotlib import cm
from IPython.display import clear_output
from matplotlib.ticker import LinearLocator, FormatStrFormatter



for angle in range(30,330,5):
    fig = plt.figure(figsize=(8,8))
    ax = fig.gca(projection='3d')

    # Plot the surface.
    X = np.arange(-2, 2, 0.01) 
    Y = np.arange(-2, 2, 0.01) 
    X, Y = np.meshgrid(X, Y)
    Z = -X**2 -Y**2
    surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                        linewidth=0, antialiased=False, alpha=0.3)
    # Plot the curve
    t = np.linspace(-1.4, 1.4, 50)
    x = t
    y = 0 + 0 * t
    z = -x**2 -y**2
    ax.plot(x, y, z, linewidth=3, color="k") 

    # Customize the z axis.
    ax.set_zlim(-10, 0.5)
    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

    # Add a color bar which maps values to colors.
    fig.colorbar(surf, shrink=0.5, aspect=5)

    # Set angle
    ax.view_init(30,angle)

    # Show the plots
    plt.tight_layout()
    plt.gca()
    clear_output(wait=True)
    plt.show()
```

Sabemos do cálculo de uma variável que $g'(0)$ representa coeficiente angular da reta tangente no ponto zero.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np
import pandas as pd
from matplotlib import cm
from IPython.display import clear_output
from matplotlib.ticker import LinearLocator, FormatStrFormatter



for angle in range(30,330,5):
    fig = plt.figure(figsize=(8,8))
    ax = fig.gca(projection='3d')
    
    # Plot the curve
    t = np.linspace(-1.4, 1.4, 50)
    x = t
    y = 0 + 0 * t
    z = -x**2 -y**2
    ax.plot(x, y, z, linewidth=3, color="k", label=r"$z=-x^2-y^2\cap y=0$") 
    
    # Plot the tangent line
    t = np.linspace(-1.4, 1.4, 50)
    x = t
    y = 0 + 0 * t
    z = 0 + 0 * t
    ax.plot(x, y, z, linewidth=3, color="g", label="Reta tangente em (0,0,0)") 

    # Plot the surface.
    X = np.arange(-2, 2, 0.01) 
    Y = np.arange(-2, 2, 0.01) 
    X, Y = np.meshgrid(X, Y)
    Z = -X**2 -Y**2
    surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                        linewidth=0, antialiased=False, alpha=0.3)
    
    # Customize the z axis.
    ax.set_zlim(-10, 0.5)
    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

    # Add a color bar which maps values to colors.
    fig.colorbar(surf, shrink=0.5, aspect=5)

    # Set the angle
    ax.view_init(30,angle)

    # Show the plots
    plt.legend()
    plt.tight_layout()
    plt.gca()
    clear_output(wait=True)
    plt.show()
```

Agora determinaremos $f_y(0,2)$. Fazenos $h(t) = f(0,t)$ basta calcular $h'(2)$

Antes, observemos que fazendo $z=f(0, t)$, temos $(x,y,z) = (0, t, -t^2)$

A reta tangente é dada por $r(t) = (0,2,-4) + t(0,1,-4)$.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np
import pandas as pd
from matplotlib import cm
from IPython.display import clear_output
from matplotlib.ticker import LinearLocator, FormatStrFormatter



for angle in range(30,330,5):
    fig = plt.figure(figsize=(8,8))
    ax = fig.gca(projection='3d')

    # Plot the surface.
    X = np.arange(-2, 2, 0.01) 
    Y = np.arange(-2, 2, 0.01) 
    X, Y = np.meshgrid(X, Y)
    Z = -X**2 -Y**2
    surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,linewidth=0, antialiased=False, alpha=0.3)
    
    t = np.linspace(-3, 3, 50)
    x = 0 + 0 * t
    y = t
    z = -t ** 2
    ax.plot(x, y, z, linewidth=3, color="k", label=r"$z=-x^2-y^2\cap x=0$") 
    
    t = np.linspace(-1.4, 1.4, 50)
    x = 0 + 0 * t
    y = 2 + t
    z = -4 - 4 * t
    ax.plot(x, y, z, linewidth=3, color="g", label="Reta tangente em (0,2,-4)") 

    # Customize the z axis.
    ax.set_zlim(-19, 0.5)
    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

    ax.view_init(30,angle)

    # Add a color bar which maps values to colors.
    fig.colorbar(surf, shrink=0.5, aspect=5)

    # configura o ângulo
    ax.view_init(30,angle)

    # mostra
    plt.legend()
    plt.tight_layout()
    plt.gca()
    clear_output(wait=True)
    plt.show()
```
