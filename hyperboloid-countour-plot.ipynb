## Curvas de nível do Hiperbolóide de uma folha

Primeiro, visualizaremos as curvas em um gráfico tridimensional.

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
from IPython.display import clear_output

for angle in range(30,330,5):
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    X = np.arange(-3, 3.1, 0.2)
    Y = np.arange(-3, 3.1, 0.2)
    X, Y = np.meshgrid(X, Y)
    Z = Y ** 2 - X ** 2 

    # Plot contour curves
    cset = ax.contour(X, Y, Z, cmap=cm.coolwarm)

    ax.clabel(cset, fontsize=9, inline=1)

    # configura o ângulo
    ax.view_init(30,angle)

    # mostra
    plt.tight_layout()
    plt.gca()
    clear_output(wait=True)
    plt.show()
```

Agora, vejamos em um gráfico bidimensional:

```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt


delta = 0.025
x = np.arange(-3, 3.1, 0.2)
y = np.arange(-3, 3.1, 0.2)
X, Y = np.meshgrid(x, y)
Z = Y ** 2 - X ** 2 

fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=8)
ax.set_title('Curvas de nível do hiperbolóide de uma folha')
plt.show()
```
