# Plotando vetores

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### Fazendo 1 vetor

```python
fig, ax = plt.subplots()

x_pos = 0
y_pos = 0
x_direct = 1
y_direct = 1

ax.quiver(x_pos,y_pos,x_direct,y_direct)

plt.show()
```

### Fazendo 2 vetores

```python
fig, ax = plt.subplots()

x_pos = [0, -0.5]
y_pos = [0, 0.5]
x_direct = [1, 0]
y_direct = [1, -1]

ax.quiver(x_pos,y_pos,x_direct,y_direct, scale=5)
ax.axis([-1.5,1.5,-1.5,1.5])

plt.show()
```

### Gradientes de uma função

```python
fig, ax = plt.subplots()

x = np.arange(0,20.2,0.5)
y = np.arange(0,20.2,0.5)

X, Y = np.meshgrid(x, y)
dx = np.cos(X)*Y
dy = np.sin(y)*Y

ax.quiver(x,y,dx,dy)

plt.show()
```

### Gradientes de uma função utilizando np.gradient

```python
x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)

X, Y = np.meshgrid(x, y)
z = X*np.exp(-X**2 -Y**2)
dx, dy = np.gradient(z)

fig, ax = plt.subplots(figsize=(7,7))
ax.quiver(X,Y,dx,dy)

ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')

plt.show()
```

### Gradiente em imagens

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

fname="images/catgray.jpg"
im = mpimg.imread(fname)

fig,ax=plt.subplots(1,1)
the_image = ax.imshow(
                im,
                zorder=0,alpha=1.0,
                cmap="Greys_r",
                origin="upper",
                interpolation="hermite",
            )
plt.colorbar(the_image)            

Y, X = np.mgrid[:im.shape[0], :im.shape[1]]
dY, dX = np.gradient(im)
ax.quiver(X, Y, dX, dY, color='r')

plt.show()
```
