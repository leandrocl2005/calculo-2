```python
def P(L, K):
    """
        L: work amount
        K: investments capital amount
        Return: production
    """
    return 1.01 * (L ** 0.75) * (K ** 0.25) 
```

```python
# Em 1910 com L=147 e K=208 a produção foi de P=159
P(147, 208)
```

```python
# Em 1920 com L=194 e K=407 a produção foi de P=231
P(194, 407)
```

Vejamos o gráfico de dispersão dos dados colhidos por Cobb e Douglas:

```python

from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
from IPython.display import clear_output

df = pd.read_csv('data/cobb-douglas.csv')

x = df['L']
y = df['K']
z = df['P']

for angle in range(30,330,3):
    fig = plt.figure()
    ax = fig.gca(projection='3d')

    ax.scatter(x, y, z, zdir='z')

    ax.set_xlim(50, 200)
    ax.set_ylim(0, 600)
    ax.set_zlim(0, 400)
    ax.set_xlabel('Quantidade de trabalho')
    ax.set_ylabel('Quantidade de investimentos')
    ax.set_zlabel('Produção')

    ax.view_init(15,angle)

    plt.tight_layout()
    plt.gca()
    clear_output(wait=True)
    plt.show()
```

Vejamos o gráfico do modelo proposto por Cobb e Douglas:

```python
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np
import pandas as pd
from matplotlib import cm
from IPython.display import clear_output
from matplotlib.ticker import LinearLocator, FormatStrFormatter

for angle in range(30,330,5):
    fig = plt.figure()
    ax = fig.gca(projection='3d')

    # Make data.
    X = np.arange(50, 200, 5) # L
    Y = np.arange(0, 600, 20) # K
    X, Y = np.meshgrid(X, Y)
    Z = P(X, Y)

    # Plot the surface.
    surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                        linewidth=0, antialiased=False)

    # Customize the z axis.
    ax.set_zlim(0, 400)
    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

    ax.view_init(30,angle)

    # Add a color bar which maps values to colors.
    fig.colorbar(surf, shrink=0.5, aspect=5)

    # configura o ângulo
    ax.view_init(30,angle)

    # mostra
    plt.tight_layout()
    plt.gca()
    clear_output(wait=True)
    plt.show()
```

```python
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np
import pandas as pd
from matplotlib import cm
from IPython.display import clear_output
from matplotlib.ticker import LinearLocator, FormatStrFormatter

for angle in range(30,330,10):
    fig = plt.figure()
    ax = fig.gca(projection='3d')

    # Make data.
    X = np.arange(50, 200, 5) # L
    Y = np.arange(0, 600, 20) # K
    X, Y = np.meshgrid(X, Y)
    Z = P(X, Y)

    # Plot the surface.
    surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                        linewidth=0, antialiased=False, alpha=0.3)

    # Customize the z axis.
    ax.set_zlim(0, 400)
    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
    
    ax.scatter(x, y, z, zdir='z', c='r')
    ax.set_xlim(50, 200)
    ax.set_ylim(0, 600)
    ax.set_zlim(0, 400)
    ax.set_xlabel('Quantidade de trabalho')
    ax.set_ylabel('Quantidade de investimentos')
    ax.set_zlabel('Produção')

    ax.view_init(30,angle)

    # Add a color bar which maps values to colors.
    fig.colorbar(surf, shrink=0.5, aspect=5)

    # configura o ângulo
    ax.view_init(0,angle)

    # mostra
    plt.tight_layout()
    plt.gca()
    clear_output(wait=True)
    plt.show()
```
