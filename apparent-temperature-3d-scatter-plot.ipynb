Gráfico de sensação térmica em relação à temperatura real e à velocidade do vento. Dados compilados pelo Serviço Nacional de Meteorologia dos Estados Unidos e pelo Serviço Meterológico do Canadá.

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
from IPython.display import clear_output

df = pd.read_csv('data/apparent-temperature.csv')

x = df['wind_speed(km/h)']
y = df['temperature(ºC)']
z = df['apparent_temperature']

for angle in range(30,330,3):
    fig = plt.figure()
    ax = fig.gca(projection='3d')

    ax.scatter(x, y, z, zdir='z')

    ax.set_xlim(-100, 200)
    ax.set_ylim(-50, 200)
    ax.set_zlim(-70, 5)
    ax.set_xlabel('Wind speed (km/h)')
    ax.set_ylabel('Temperature (C)')
    ax.set_zlabel('Apparent Temperature')

    ax.view_init(15,angle)

    plt.tight_layout()
    plt.gca()
    clear_output(wait=True)
    plt.show()
```
